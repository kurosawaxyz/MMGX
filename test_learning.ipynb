# Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read learning record

```python
dataset = './dataset/bbbp/GIN_AR_0_functional/2022-Dec-19-16:50:16'
my_cols = [str(i) for i in range(53)]
df = pd.read_csv(dataset+'/logs/exp-001_seed-42_fold-4.log', header=None, sep="\s+|,|:|\[|\]", skipfooter=1, engine='python')
```

```python
epoch = list(df[5])
train_loss = list(df[9])
val_loss = list(df[13])
val_eval = list(df[17])
```

```python
data_dict = {'epoch': epoch, 'train_loss': train_loss, 'val_loss': val_loss, 'val_eval': val_eval}
df_plot = pd.DataFrame(data=data_dict)
df_plot = df_plot.set_index('epoch')
```

```python
df_plot
```

# Visualization

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.lineplot(data=df_plot)
```

```python
data_dict = {'epoch': epoch, 'val_eval': val_eval}
df_plot = pd.DataFrame(data=data_dict)
df_plot = df_plot.set_index('epoch')
sns.set(rc = {'figure.figsize':(15,8)})
sns.lineplot(data=df_plot)
```
